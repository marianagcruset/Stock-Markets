```python
#Str to number
btc_df["open"] = pd.to_numeric(btc_df["open"])
btc_df["high"] = pd.to_numeric(btc_df["high"])
btc_df["low"] = pd.to_numeric(btc_df["low"])
btc_df["close"] = pd.to_numeric(btc_df["close"])
btc_df["adjclose"] = pd.to_numeric(btc_df["adjclose"])
btc_df["volume"] = pd.to_numeric(btc_df["volume"])
```

```python
def label_dataset(t_horizon):
    for row in btc_df.index:
        if row < len(btc_df.index)-1:
            
            # Calculate the maximum value for the next t_horizon rows of high values.
            btc_df.loc[row, f"Max_{t_horizon}"] = max(btc_df.loc[(row + 1):(row + t_horizon), "high"]) 
            
            # Calculate the minimum value for the next t_horizon rows of low values.
            btc_df.loc[row, f"Min_{t_horizon}"] = min(btc_df.loc[(row+1):(row+t_horizon), "low"])
            
            # Calculate the percentage gain from going long if exits at optimal price next 3 days
            btc_df.loc[row, f"Max_{t_horizon}_Percent_Increase"] = ((btc_df.loc[row, f"Max_{t_horizon}"] - 
                                                                     btc_df.loc[row, "close"]) / btc_df.loc[row, "close"])
           
            # Calculate the percentage gain from going Short if exits at optimal price next 3 days 
            btc_df.loc[row, f"Min_{t_horizon}_Percent_Increase"] = ((btc_df.loc[row, "close"] - 
                                                                     btc_df.loc[row, f"Min_{t_horizon}"]) / btc_df.loc[row, "close"])
            
            # What was the optimal move? (SHORT or LONG)
            btc_df.loc[row, f"Choice_{t_horizon}"] = "LONG" if btc_df.loc[row, f"Max_{t_horizon}_Percent_Increase"] > btc_df.loc[row, f"Min_{t_horizon}_Percent_Increase"] else "SHORT"
            
            # What is the optimate gain?
            btc_df.loc[row, f"Gain_{t_horizon}"] = btc_df.loc[row, f"Max_{t_horizon}_Percent_Increase"] if btc_df.loc[row, f"Max_{t_horizon}_Percent_Increase"] > btc_df.loc[row, f"Min_{t_horizon}_Percent_Increase"] else btc_df.loc[row, f"Min_{t_horizon}_Percent_Increase"]

label_dataset(3)
label_dataset(6)
```

```python
def day_of_week_standard_form(string_date):
    date_object = dt.datetime.strptime(string_date, "%m/%d/%Y")
    return date_object.dayofweek

# btc_df["Weekday"] = btc_df["Date"].map(day_of_week_standard_form)
maxbtc_df["date"] = maxbtc_df["date"].map(lambda x: dt.datetime.strptime(x, "%m/%d/%Y"))
maxbtc_df["Weekday"] = maxbtc_df["date"].dt.dayofweek
```

```python
# Exponential Moving Averages indicators
ema_6 = ta.trend.ema_indicator(close = maxbtc_df["close"], window = 6)
ema_12 = ta.trend.ema_indicator(close = maxbtc_df["close"], window = 12)
ema_24 = ta.trend.ema_indicator(close = maxbtc_df["close"], window = 24)

# RSI indicators
rsi_14 = ta.momentum.RSIIndicator(maxbtc_df["close"], window = 14)
rsi_7 = ta.momentum.RSIIndicator(maxbtc_df["close"], window = 7)
rsi_28 = ta.momentum.RSIIndicator(maxbtc_df["close"], window = 28)


# Create new columns based on the indicators
maxbtc_df["ema_6"] = ema_6
maxbtc_df["ema_12"] = ema_12
maxbtc_df["ema_24"] = ema_24
maxbtc_df["p_ema_6"] = maxbtc_df["ema_6"] / maxbtc_df["close"]
maxbtc_df["p_ema_12"] = maxbtc_df["ema_12"] / maxbtc_df["close"]
maxbtc_df["p_ema_24"] = maxbtc_df["ema_24"] / maxbtc_df["close"]
maxbtc_df["rsi_7"] = rsi_7.rsi()
maxbtc_df["rsi_14"] = rsi_14.rsi()
maxbtc_df["rsi_28"] = rsi_28.rsi()
```

```python
emabtc_df.dropna(inplace = True)
```

```python
print(__doc__)
RANDOM_STATE = 123
# Generate a binary classification dataset.
# X, y = make_classification(n_samples=500, n_features=25,
#                           n_clusters_per_class=1, n_informative=15,
#                           random_state=RANDOM_STATE)

X = emabtc_df[["ema_6", "ema_12", "ema_24", "p_ema_6", "p_ema_12", "p_ema_24", "rsi_7", "rsi_14", "rsi_28", "Weekday"]]

y = emabtc_df["Choice_3"]

# NOTE: Setting the 'warm_start' construction parameter to 'True' disables
# support for parallelized ensembles but is necessary for tracking the OOB
# error trajectory during training.
ensemble_clfs = [
    ("RandomForestClassifier, max_features='auto'",
        RandomForestClassifier(warm_start=True, oob_score=True,
                               max_features="auto",
                               random_state=RANDOM_STATE)),
    ("RandomForestClassifier, max_features='log2'",
        RandomForestClassifier(warm_start=True, max_features='log2',
                               oob_score=True,
                               random_state=RANDOM_STATE)),
    ("RandomForestClassifier, max_features=None",
        RandomForestClassifier(warm_start=True, max_features=None,
                               oob_score=True,
                               random_state=RANDOM_STATE))
]

# Map a classifier name to a list of (<n_estimators>, <error rate>) pairs.
error_rate = OrderedDict((label, []) for label, _ in ensemble_clfs)
error_rate_test = OrderedDict((label, []) for label, _ in ensemble_clfs)

# Range of 'n_estimators' values to explore.
min_estimators = 15
max_estimators = 500
time_step = 5
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

for label, clf in ensemble_clfs:
    for i in range(min_estimators, max_estimators, time_step):
        clf.set_params(n_estimators=i)
        clf.fit(X_train, y_train)

        # Record the OOB error for each `n_estimators=i` setting.
        oob_error = 1 - clf.oob_score_
        error_rate[label].append((i, oob_error))
        oob_error_test = 1 - clf.score(X_test, y_test)
        error_rate_test[label].append((i, oob_error_test))
```

```python
fig, axs = plt.subplots(2, figsize=(15, 8))
fig.suptitle('RF Training vs Testing')

# Generate the "OOB error rate" vs. "n_estimators" plot.
for label, clf_err in error_rate_test.items():
    xs, ys = zip(*clf_err)
    axs[0].plot(xs, ys, label=label)
    
# Generate the "OOB error rate" vs. "n_estimators" plot.
for label, clf_err in error_rate.items():
    xs, ys = zip(*clf_err)
    axs[1].plot(xs, ys, label=label)

plt.xlim(min_estimators, max_estimators)
plt.xlabel("n_estimators")
axs[1].set_ylabel("Error Training")
axs[0].set_ylabel("Error Test")

# plt.legend(loc="outside")
plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
# plt.ylim(0,0.40)
plt.show()

```
